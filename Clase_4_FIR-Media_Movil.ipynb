# Clase 4: Filtro Media movil y primera diferencia

```python
# Importar las librerias necesarias
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
x = list(range(1, 45))

n = 44  # Nmero de elementos
y = []
for i in range(n):
    numero_aleatorio = random.randint(30, 60)
    y.append(numero_aleatorio)
    
plt.figure(figsize=(15,6))
plt.plot(x,y,linewidth=3)
plt.title("Seal aleatoria",fontsize=25)
plt.grid()
plt.show()
```

```python
L=5
coefs=np.ones(L)/float(L)
senal_smooth=np.convolve(y, coefs, mode='same')

L1=10
coefs1=np.ones(L1)/float(L1)
senal_smooth1=np.convolve(y, coefs1, mode='same')

L2=15
coefs2=np.ones(L2)/float(L2)
senal_smooth2=np.convolve(y, coefs2, mode='same')

L3=30
coefs3=np.ones(L3)/float(L3)
senal_smooth3=np.convolve(y, coefs3, mode='same')

plt.figure(figsize=(18,18))

plt.subplot(411)
plt.plot(x,senal_smooth,linewidth=3)
plt.title("Seal con filtro L=5",fontsize=25)
plt.grid()

plt.subplot(412)
plt.plot(x,senal_smooth1,linewidth=3)
plt.title("Seal con filtro L=10",fontsize=25)
plt.grid()

plt.subplot(413)
plt.plot(x,senal_smooth2,linewidth=3)
plt.title("Seal con filtro L=15",fontsize=25)
plt.grid()

plt.subplot(414)
plt.plot(x,senal_smooth3,linewidth=3)
plt.title("Seal con filtro L=30",fontsize=25)
plt.grid()

plt.subplots_adjust(hspace=1)
```

```python
difcoefs=[1, -1]
senal_diff=np.convolve(y, difcoefs, mode='same')

num=[1, -1]
den=[1, 0]

plt.figure(figsize=(18,8))

plt.subplot(211)
plt.plot(x,y,linewidth=3)
plt.title("Seal",fontsize=25)
plt.grid()

plt.subplot(212)
plt.plot(x,senal_diff,linewidth=3)
plt.title("Seal con filtro",fontsize=25)
plt.grid()

plt.subplots_adjust(hspace=1)
```
