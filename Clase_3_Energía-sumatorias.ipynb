```python
# Importar las librerias necesarias
import numpy as np
import math
import matplotlib.pyplot as plt
from functools import reduce
```

```python
n=2
fs = (n+1)*22 # Frecuencia de muestreo
t=np.arange(0, 1.0, 5/fs) # Vector de tiempo
x1 = 1/t**2

plt.plot(t,x1,label="f1=3 Hz")
plt.xlabel('Tiempo',fontsize=18)
plt.ylabel('Amplitud',fontsize=18)
plt.legend()
plt.grid()
plt.show()

print("Frecuencia de muestreo: {} Hz".format(fs))
```

```python
# primeros 40 valores de k
k = np.arange(40)
def fn(k):
    t = 10  # algun t
    return [reduce(lambda elem, acc: elem + acc, [(1/n)**2 for j in range(k_i + 1)]) for k_i in k]

y = fn(k)

plt.plot(k, y, color="r")
plt.xlabel('k')
plt.ylabel('y')
plt.title('Serie de Fourier')
plt.grid()
plt.show()
```
