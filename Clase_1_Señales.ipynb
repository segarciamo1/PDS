# Clase 1. Se√±ales.

```python
# Importar las librerias necesarias
import numpy as np
import matplotlib.pyplot as plt
```

```python
w  = 2*np.pi/12
fx = lambda n: np.cos(w*n)

n0 = 0  # intervalo en un periodo
muestras = 12+1

ni = np.arange(n0,n0+muestras,1)

senal = fx(ni)

np.set_printoptions(precision=4)

plt.stem(ni, senal)

plt.xlabel('n')
plt.ylabel('Se√±al Discretizada')
plt.grid()

plt.show()
```

```python
t = np.linspace(0, 2*np.pi, 15) # Vector de tiempo
signal = np.sin(t)              # Se√±al analogica
```

```python
# Grafica de la se√±al anal√≥gica y digital
plt.figure()
plt.plot(t, signal, 'b', label = 'Se√±al anal√≥gica')
plt.stem(t, signal, 'r', label = 'Se√±al digitalizada')
plt.xlabel('Tiempo')
plt.ylabel('Amplitud')
plt.grid()
plt.legend()
plt.show()
```
