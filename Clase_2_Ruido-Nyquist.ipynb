# Clase 2: Ruido, Nyquist

```python
# Importar las librerias necesarias
import numpy as np
import matplotlib.pyplot as plt
```

```python
n=2
fs = (n+1)*22 # Frecuencia de muestreo
senal_1 = (n+1)
senal_2 = (2*fs)+(n+2)

t=np.arange(0, 1.0, 1.0/fs) # Vector de tiempo

x1 = np.sin(2*np.pi*senal_1*t)
x2 = np.sin(2*np.pi*senal_2*t)

plt.plot(t,x1,label="f1=3 Hz")
plt.plot(t,x2,label="f2=136 Hz")
plt.xlabel('Tiempo',fontsize=18)
plt.ylabel('Amplitud',fontsize=18)
plt.legend()
plt.grid()
plt.show()

print("Frecuencia de muestreo: {} Hz".format(fs))
print("Frecuencia señal 1: {} Hz".format(senal_1))
print("Frecuencia señal 2: {} Hz".format(senal_2))
```

```python
t = np.linspace(0, 2*np.pi, 50)     # Vector de tiempo
signal = np.sin(t)                  # Señal
noise = np.random.randn(50)         # ruido blanco
signal_with_noise = signal + noise  # Se añade el ruido a la señal
```

```python
# Grafica de las señales con y sin ruido
plt.figure()
plt.plot(t, signal, 'b', label = 'Señal sin ruido')
plt.plot(t, signal_with_noise, 'r', label = 'Señal con ruido')
plt.xlabel('Tiempo')
plt.ylabel('Amplitud')
plt.grid()
plt.legend()
plt.show()
```
